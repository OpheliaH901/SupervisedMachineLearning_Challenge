```python
%matplotlib inline
import numpy as np
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import balanced_accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.utils.random import sample_without_replacement
from sklearn.utils import resample
```

```python
train_2019df = pd.read_csv(Path('Resources/2019loans.csv')).drop(["Unnamed: 0"], axis=1)
test_2020df = pd.read_csv(Path('Resources/2020Q1loans.csv')).drop(["Unnamed: 0"], axis=1)
train_2019df["year"]=2019
test_2020df["year"]=2020
```

```python
merge_df=pd.concat([train_2019df, test_2020df])
merge_df
```

```python
# Convert categorical data to numeric and separate target feature for testing data
x_predummies = merge_df.drop(["loan_status"], axis=1)
x_dummies = pd.get_dummies(x_predummies)
x_dummies
```

```python
y_test = []
for label in test_2020df["loan_status"].values:
    if label == 'low_risk':
        y_test.append(0)
    else:
        y_test.append(1)     
```

```python
y_train = []
for label in train_2019df["loan_status"].values:
    if label == 'low_risk':
        y_train.append(0)
    else:
        y_train.append(1) 
```

```python
#split the data by year
x_train= x_dummies[x_dummies["year"]==2019]
x_test=x_dummies[x_dummies["year"]==2020]
```


# Train the models using unscaled data

```python
# Train the Balanced Random Forest Classifier model on the unscaled data and 
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
brf.fit(x_train, y_train)

# print the model score
predictions=brf.predict(x_test)
balanced_accuracy_score(y_test, predictions)
```

```python
classifier = LogisticRegression()
classifier.fit(x_train, y_train)

# print the model score
predictions=classifier.predict(x_test)
balanced_accuracy_score(y_test, predictions)
```

```python
# Train a Random Forest Classifier model and print the model score

clf = RandomForestClassifier(random_state=1, n_estimators=100).fit(x_train, y_train)
print(f'Training Score: {clf.score(x_train, y_train)}')
```

# Train the models using scaled data

```python
#scale the data (example of standardizing your data - crime in city for every 1000 ppl, not based on exact values)
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
```

```python
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
brf.fit(x_train_scaled, y_train)

# print the model score
predictions=brf.predict(x_test)
balanced_accuracy_score(y_test, predictions)
```

```python
classifier = LogisticRegression()
classifier.fit(x_train_scaled, y_train)

# print the model score
predictions=classifier.predict(x_test)
balanced_accuracy_score(y_test, predictions)
```

```python
# Train a Random Forest Classifier model and print the model score

classifier = RandomForestClassifier()
classifier.fit(x_train_scaled, y_train)

# print the model score
predictions=classifier.predict(x_test)
balanced_accuracy_score(y_test, predictions)
```
